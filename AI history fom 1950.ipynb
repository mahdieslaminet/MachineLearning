# تاریخچه هوش مصنوعی

<span style='font-family:XB Nazanin;font-size:19px'> 
از هوش مصنوعی در مراقبت های بهداشتی، هوش مصنوعی در آموزش، هوش مصنوعی در تولید، هوش مصنوعی در حقوق، امور مالی و حتی هوش مصنوعی در سیاست، در زمان های غالب، هوش مصنوعی در همه جا وجود دارد. با این حال، در نظر گرفتن هوش مصنوعی به عنوان یک پدیده کاملاً مدرن، تصوری نادرست خواهد بود.   

<span style='font-family:XB Niloofar;font-size:18px'>


+ 1950 : برنامه نویسی یک کامپیوتر برای بازی شطرنج، اولین مقاله در مورد ایجاد یک برنامه کامپیوتری شطرنج بازی توسط کلود شانون منتشر شد که به عنوان "پدر نظریه اطلاعات" شناخته می شد.  
+ 1950 : یکی از نقاط عطف تاریخ هوش مصنوعی آلن تورینگ بود که مقاله خود را در مورد "ماشین آلات محاسباتی و هوش" در سال جاری منتشر کرد. این مقاله ایده «بازی تقلید» را بررسی کرد که بعداً به «آزمون تورینگ» تغییر نام داد، پیشنهادی که ظرفیت یک ماشین برای تفکر مانند یک انسان را بررسی می‌کرد. این آزمایش به عنصری جدایی ناپذیر در حوزه هوش مصنوعی تبدیل شد  
+ 1952 : برنامه ای برای بازی چکرز توسط آرتور ساموئل ساخته شد. این اولین برنامه بازی بود که توانایی رقابت با بازیکنان انسانی در بازی چکرز را داشت.  
+ 1955 : اولین برنامه کامپیوتری هوش مصنوعی - نظریه منطق منطق امسال توسط آلن نیول (محقق)، هربرت سایمون (اقتصاددان) و کلیف شاو (برنامه نویس) نوشته شد که در نهایت 38 قضیه از 52 قضیه اول در وایتهد و راسل Principia Mathematica را اثبات کرد .  
+ 1955 : پیشنهادی توسط جان مک کارتی، دانشمند کامپیوتر آمریکایی و تیمش برای برگزاری کارگاه آموزشی "هوش مصنوعی" ارائه شد.
+ 1956 : پس از پایان کارگاه، اصطلاح هوش مصنوعی به طور رسمی با حسن نیت جان مک کارتی در کنفرانس کالج دارتموث متولد شد  
+ 1959 : اصطلاح «یادگیری ماشینی» توسط آرتور ساموئل به وجود آمد که او روی برنامه‌نویسی رایانه‌ای برای انجام یک بازی شطرنج کار می‌کرد که بتواند با بازیکنان انسانی رقابت کند.

<span style='font-family:XB Niloofar;font-size:19px'>
در دهه شصت ، مشخصه این دهه افزایش توسعه خودکارها و ربات ها، زبان های برنامه نویسی جدید، تحقیقات و فیلم هایی بود که موجودات با هوش مصنوعی را به تصویر می کشیدند.

+ 1961: در این سال، یک ربات صنعتی دهه 1950 به نام Unimate که توسط جورج دیول ساخته شد، به عنوان اولین ربات در خط مونتاژ جنرال موتورز نیوجرسی ظاهر شد. وظایف ربات شامل وظایفی است که انجام آنها توسط انسان خطرناک تلقی می شود، مانند حمل و نقل ریخته گری از خط مونتاژ.  
+ 1961: SAINT (یکپارچه ساز خودکار نمادین) توسط جیمز اسلاگل، دانشمند کامپیوتر و پروفسور توسعه داده شد. این برنامه ای است که مسائل ادغام نمادین را در محاسبات سال اول حل می کند  
+ 1964: یک برنامه هوش مصنوعی اولیه به نام STUDENT در Lisp توسط Daniel G. Bobrow، دانشمند کامپیوتر آمریکایی نوشته شد. این برنامه برای خواندن و حل مسائل جبر ساخته شده بود  
+ 1966: از قضا، اولین ربات چت Eliza، یک برنامه کامپیوتری پردازش زبان طبیعی، که توسط پروفسور MIT جوزف وایزنبام توسعه یافته بود، در واقع به عنوان یک تقلید به جای چیزی که مردم می توانستند با آن ارتباط برقرار کنند، ساخته شده بود. با این حال مردم شروع به نشان دادن ارتباطات عاطفی واقعی با آن کردند. اگرچه الیزا از طریق متن تعامل داشت و قادر به یادگیری از طریق تعاملات انسانی نبود، او نقش مهمی در از بین بردن موانع ارتباطی بین انسان و ماشین داشت.  
+ 1966: این سالی بود که پروژه ساخت اولین ربات متحرک به نام Shakey را آغاز کرد. این پروژه از سال 1966 تا 1972 به طول انجامید. این پروژه به عنوان رویکردی برای پیوند دادن زمینه های مختلف هوش مصنوعی با ناوبری و بینایی کامپیوتری در نظر گرفته شد. این ربات در حال حاضر در موزه تاریخ کامپیوتر اقامت دارد.  
+ 1968: این سالی بود که فیلم علمی تخیلی محبوب 2001: یک ادیسه فضایی اکران شد. این فیلم به کارگردانی استنلی کوبریک، فیلمساز، HAL (رایانه الگوریتمی برنامه ریزی شده اکتشافی)، یک سفینه فضایی دیسکاوری وان با انرژی هسته ای مجهز به هوش مصنوعی بود. ابرکامپیوتر آنبورد شباهت خاصی به دستیارهای صوتی مدرن امروزی ما مانند سیری و الکسا داشت  
+ 1968: SHRDLU، یک برنامه کامپیوتری اولیه به زبان طبیعی توسط تری وینوگراد، استاد علوم کامپیوتر در دانشگاه استنفورد ایجاد شد  


<span style='font-family:XB Niloofar;font-size:19px'>
در دهه هفتاد ، این دهه با پیشرفت هایی در رابطه با روبات ها سروکار داشت. با این حال، با چالش‌های مختلفی مانند کاهش حمایت دولت از تحقیقات در زمینه هوش مصنوعی رو به رو شد.

+ 1970: این سالی بود که اولین ربات انسان نما به نام WABOT-1 در ژاپن در دانشگاه Waseda ساخته شد. اندام متحرک و توانایی مشاهده و مکالمه از جمله ویژگی هایی بود که در آن گنجانده شده بود  
+ 1973: پیشرفت در هوش مصنوعی در این مرحله زمانی که جیمز لایتهیل، ریاضیدان کاربردی، وضعیت تحقیقات هوش مصنوعی را به شورای علمی بریتانیا اعلام کرد و مدعی شد که هنوز هیچ یک از اکتشافات انجام شده تاثیر مورد انتظار را ایجاد نکرده است، رو به نزول گذاشت. این منجر به کاهش قابل توجه دولت بریتانیا از تحقیقات هوش مصنوعی شد.  
+ 1979: Stanford Cart، یک ربات متحرک مجهز به تلویزیون کنترل از راه دور که در سال 1961 ساخته شد، با موفقیت از اتاقی پر از صندلی در غیاب دخالت انسان در مدت 5 ساعت عبور کرد که آن را به یکی از اولیه ترین وسایل نقلیه خودمختار تبدیل کرد.

<span style='font-family:XB Niloofar;font-size:19px'>
در دهه هشتاد ، زمستان هوش مصنوعی (دوره‌ای که با سطح پایین علاقه و بودجه در زمینه هوش مصنوعی مشخص می‌شود) در بخش خاصی از این دهه پیش می‌آید که بعداً با از سرگیری بودجه دولت بریتانیا به عنوان تلاشی برای رقابت با تلاش‌های ژاپنی احیا شد.

+ 1980: توسعه WABOT-2 در دانشگاه Waseda انجام شد و انسان نما را قادر ساخت تا با مردم ارتباط برقرار کند، موسیقی را بخواند و موسیقی را بر روی یک ارگ الکترونیکی پخش کند.  
+ 1981: حدود 850 میلیون دلار به پروژه نسل پنجم کامپیوتر از طرف وزارت تجارت و صنعت بین‌المللی ژاپن اختصاص یافت، هدف این پروژه توسعه کامپیوتر هایی است که می‌توانند تعاملات را اجرا کنند، زبان‌ها را ترجمه کنند، تصاویر را تفسیر کنند و مانند انسان‌ها تجزیه و تحلیل کنند.
+ 1984: Electric Dreams، فیلمی به کارگردانی استیو بارون منتشر شد، داستانی جالب درباره مثلث عشقی بین یک مرد، زن و یک کامپیوتر شخصی به نام "ادگار"  
+ 1984: هشدار در مورد زمستان هوش مصنوعی در آینده در انجمن پیشرفت هوش مصنوعی (AAAI) توسط راجر شانک و ماروین مینسکی مطرح شد. این هشدار ظرف 3 سال آینده به وقوع پیوست.  
+ 1986: این سالی بود که اولین خودروی بدون راننده، یک ون مرسدس بنز، مجهز به حسگرها و دوربین‌ها که تحت هدایت ارنست دیکمن ساخته شد، در خیابان‌های خالی تا 55 مایل در ساعت رانندگی کرد.
+ 1988: Judea Pearl، دانشمند کامپیوتر، "استدلال احتمالی در سیستم های هوشمند" را در این سال منتشر کرد. همچنین اعتبار ابداع شبکه‌های بیزی، فرمالیسم ریاضی برای تعریف مدل‌های احتمال پیچیده و الگوریتم‌های اولیه برای استنتاج در این مدل‌ها به او داده شد.  
+ 1988: Rollo Carpenter، برنامه نویس Jabberwacky را با هدف شبیه سازی چت طبیعی انسان به شیوه ای جذاب ساخت. این یکی از رویکردهای اولیه برای تولید هوش مصنوعی از طریق تعامل انسانی بود  


<span style='font-family:XB Niloofar;font-size:19px'>
در دهه نود ، پایان هزاره با طیف وسیعی از پیشرفت‌ها مشخص می‌شود که به تدریج حوزه هوش مصنوعی را متحول کرد

+ 1993: کتاب «فیل ها شطرنج بازی نمی کنند» توسط رادنی بروکس منتشر شد. این کتاب رویکرد جدیدی را برای هوش مصنوعی در توسعه سیستم‌های هوشمند از ابتدا و بر اساس تعامل فیزیکی مداوم با محیط پیشنهاد می‌کند.  
+ 1995: A.L.I.C.E (Artificial Linguistic Internet Computer Entity) توسط ریچارد والاس، دانشمند کامپیوتر ساخته شده است که از ELIZA Weizenbaum الهام گرفته شده است. افزودن مجموعه داده های نمونه زبان طبیعی در A.L.I.C.E این دو را متمایز کرد.  
+ 1997: یورگن اشمیدهابر و سپ هوکرایتر، حافظه کوتاه مدت بلند مدت (LSTM) را پیشنهاد کردند، نوعی از معماری شبکه عصبی بازگشتی (RNN) که در حال حاضر برای تشخیص گفتار و دست خط استفاده می شود.
+ 1997: این سالی بود که کامپیوتر شطرنج بازی IBM Deep Blue به عنوان اولین سیستمی که در برابر قهرمان شطرنج جهان پیروز شد، ظاهر شد.  
+ 1998: Furby، اولین ربات اسباب بازی خانگی یا خانگی توسط دیو همپتون و کالب چونگ اختراع شد.  
+ 1998: ربات انسان نمای رسا "Kismet" توسط پروفسور سینتیا بریزیل MIT ساخته شده است. این روباتی است که می تواند احساسات را از طریق چهره خود شناسایی و شبیه سازی کند. ساختار این ربات شبیه صورت انسان بود که مجهز به چشم، لب، پلک و ابرو بود  
+ 1999: با پیروی از Furby، سونی AIBO (ربات هوش مصنوعی) را معرفی کرد که برای یادگیری از طریق تعامل با محیط و صاحبان انسان طراحی شده بود. این ربات توانایی درک و پاسخگویی به بیش از 100 فرمان صوتی را داشت.

<span style='font-family:XB Niloofar;font-size:19px'>
هوش مصنوعی از 2000 تا 2010 ، اگرچه هزاره با وحشت ناشی از باگ Y2K در همان ابتدای سال 2000 همراه شد، پس از حل آن، روند پیشرفت در دنیای هوش مصنوعی مانند ققنوس افزایش یافت

+ 2000: یک ربات انسان نما با هوش مصنوعی به نام آسیمو توسط هوندا منتشر شد. این ربات قادر است به سرعت انسان راه برود و سینی هایی را به مشتریان رستوران تحویل دهد  
+ 2001: فیلم علمی تخیلی استیون اسپیلبرگ A.I. هوش مصنوعی امسال منتشر شد. این فیلم حول محور دیوید می چرخید، یک اندروید کودکانه که منحصراً با توانایی عشق ورزیدن برنامه ریزی شده بود  
+ 2002: رومبا محبوب توسط i-Robot منتشر شد که یک جاروبرقی رباتی مستقل بود که در حین دور زدن موانع تمیز می کند.  
+ 2004: این سالی بود که مریخ نوردهای اکتشافی رباتیک ناسا یعنی Spirit و Opportunity در غیاب مداخله انسانی در سطح مریخ حرکت کردند.  
+ 2004: من، ربات، فیلمی در سال 2035 که در آن ربات های انسان نما به بشریت خدمت می کردند، به کارگردانی الکس پرویاس کارگردان استرالیایی اکران می شود.
+ 2006: اصطلاح "ماشین خوانی" توسط اورن اتزیونی، میشل بانکو و مایکل کافارلا ابداع شد و آن را به عنوان درک متن مستقل بدون نظارت تعریف کردند.  
+ 2007: ImageNet، یک پایگاه داده بزرگ حاوی تصاویر حاشیه نویسی شده توسط فی فی لی، استاد علوم کامپیوتر، و همکارانش جمع آوری شد. هدف این پایگاه داده کمک به تحقیقات نرم افزاری تشخیص اشیا بود.  
+ 2009: یک ماشین بدون راننده مخفیانه توسط گوگل ساخته شد. در سال 2014، این خودرو موفق شد آزمون خودران نوادا را پشت سر بگذارد.

نکته ای کوتاه از هوش مصنوعی حاضر  


<span style='font-family:XB Niloofar;font-size:19px'>
از این نقطه به بعد، هوش مصنوعی در زندگی روزمره ما به طور کامل خود را در تمام وظایف روزمره ما جا داده است. با تسلط بر هر گوشه ای از وجود خود، به نقطه ای رسیده ایم که حتی ممکن است این فناوری را بدیهی فرض کنیم. ما چند مورد از نکات برجسته در این زمینه را که در دهه اخیر رخ داده است، فهرست کرده ایم.

+ 2010: مایکروسافت ، کینکت را برای ایکس باکس 360 منتشر کرد که اولین دستگاه بازی بود که حرکت بدن انسان را از طریق دوربین سه بعدی و تشخیص مادون قرمز ردیابی می کرد.  
+ 2011: دو نفر از خطرات سابق! قهرمانان کن جنینگ و برد راتر در یک بازی تلویزیونی به واتسون از IBM، یک کامپیوتر پاسخگوی سوالات به زبان طبیعی، شکست خوردند  
+ 2011: اپل سیری را منتشر کرد که یک دستیار شخصی داخلی با کنترل صوتی است که برای کاربران اپل در دسترس است. این دستیار صوتی یک رابط کاربری با زبان طبیعی برای درک، مشاهده، پاسخ دادن و پیشنهاد دادن چیزها به کاربران انسانی با تطبیق با دستورات صوتی به کار می‌برد.  
+ 2012: یک شبکه عصبی عظیم مجهز به 16000 پردازنده توسط محققان گوگل جف دین و اندرو نگ آموزش دیدند تا تصاویر گربه را بدون هیچ گونه اطلاعات پس زمینه با ارائه 10 میلیون تصویر بدون برچسب موجود در ویدئوهای یوتیوب شناسایی کنند  
+ 2013: یک برنامه کامپیوتری به نام یادگیرنده تصویر بی پایان (NEIL) که در دانشگاه کارنگی ملون قرار دارد، امسال منتشر شد. این برنامه 24/7 کار می کند و اطلاعات مربوط به تصاویری را که در اینترنت کشف می کند، یاد می گیرد  
+ 2014: الکسای افسانه ای توسط آمازون منتشر شد که یک دستیار خانگی است که به بلندگوهای هوشمند تبدیل می شود و به عنوان دستیار شخصی عمل می کند.  
+ 2015: نامه ای سرگشاده توسط ایلان ماسک، استیون هاوکینگ و استیو وزنیاک در میان 3000 نفر دیگر امضا شد که در آن خواستار ممنوعیت توسعه و استفاده از سلاح های خودمختار برای اهداف جنگی شدند.  
+ 2016: Hanson Robotics ربات انسان نمای سوفیا را ساخت که اولین شهروند رباتی است. وجه تمایز او از پیشینیانش شباهت او به یک انسان واقعی، با توانایی او در دیدن، ایجاد حالات چهره و برقراری ارتباط از طریق هوش مصنوعی بود.  
+ 2016: Google Home توسط گوگل منتشر شد، یک بلندگوی هوشمند که هوش مصنوعی را برای خدمت به عنوان «دستیار شخصی» برای کمک به کاربران در وظایفی مانند یادآوری وظایف، ایجاد قرار ملاقات و یادگیری اطلاعات با استفاده از صدا، به کار می‌گیرد.  
+ 2017: دو "عامل گفتگو" (ربات‌های گفتگو) توسط آزمایشگاه تحقیقاتی هوش مصنوعی فیس‌بوک (چت ربات) آموزش می‌بینند تا از طریق یادگیری ماشین مذاکره کنند. با این حال، همانطور که چت ربات ها با هم تعامل داشتند، از زبان انسانی منحرف شدند و زبان ارتباطی خود را اختراع کردند.  
+ 2018: Alibaba یک مدل هوش مصنوعی ساخته است که در آزمون خواندن و درک مطلب دانشگاه استنفورد امتیاز بهتری نسبت به انسان کسب کرده است. پردازش زبان علی بابا در مجموعه ای از 100000 سوال، امتیاز 82.44 در مقابل 82.30 را کسب کرد.  
+ 2018: گوگل BERT (نمایش رمزگذار دوطرفه از Transformers) را توسعه داد. این فناوری به هر کسی اجازه می دهد تا سیستم پاسخگویی به سوالات پیشرفته خود را آموزش دهد  
+ 2020: OpenAI GPT-3، اولین بار در ماه مه 2020 معرفی شد و در ژوئن 2020 در آزمایش بتا قرار گرفت. این مدل زبان متنی را با استفاده از الگوریتم‌هایی تولید می‌کند که از قبل آموزش داده شده‌اند، یعنی قبلاً داده‌هایی را که برای اجرای کار نیاز دارند تغذیه شده‌اند.
